```python
import os


def offline_chat(messages):
    # Offline fallback so the notebook runs without any paid API keys.
    user_msg = next((m.get("content", "") for m in reversed(messages) if m.get("role") == "user"), "")
    if "usa" in user_msg.lower() or "united states" in user_msg.lower():
        return (
            "(offline demo) Here are quick facts about the United States:\n"
            "- Capital: Washington, D.C.\n"
            "- 50 states; federal republic\n"
            "- Largest cities include New York City, Los Angeles, Chicago\n"
            "- Major regions: Northeast, South, Midwest, West\n"
            "\n"
            "Ask a more specific question (history, economy, geography, culture) and I can tailor this offline summary."
        )
    return f"(offline demo) You asked: {user_msg}\n\nSet OPENAI_API_KEY to use a real model-backed answer."


def chat(messages, model=None):
    # Uses OpenAI only if you provide a working key; otherwise falls back.
    api_key = os.getenv("OPENAI_API_KEY")
    if not api_key:
        return offline_chat(messages)

    from openai import OpenAI

    model = model or os.getenv("OPENAI_MODEL", "gpt-3.5-turbo")
    try:
        client = OpenAI(api_key=api_key)
        completion = client.chat.completions.create(model=model, messages=messages)
        return completion.choices[0].message.content
    except Exception as e:
        print(f"OpenAI API call failed ({type(e).__name__}: {e}). Falling back to offline demo.")
        return offline_chat(messages)
```

##### Create Chat Completion Object

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Tell me about USA"},
]

reply = chat(messages)
print(reply)
```
