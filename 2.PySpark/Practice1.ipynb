````python
from pyspark import SparkConf, SparkContext
````

````python
conf = SparkConf()
conf.setMaster("local")
conf.setAppName('spark-basic')
sc = SparkContext(conf=conf)
````

````python
import numpy as np
def mod(x):
    return (x,np.mod(x,2))
rdd = sc.parallelize(range(1000)).map(mod).take(10)
print(rdd)
````

getOrCreate() reuses existing SparkContext instead of creating a new one.

````python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("spark-basic") \
    .master("local") \
    .getOrCreate()

sc = spark.sparkContext
````

````python
# creating an RDD
rdd = sc.parallelize(range(1000)).map(mod).take(10)
````

````python
print(rdd)
````

````python
## creating an RDD using List
values = [1, 2, 3, 4, 5]
rdd2 = sc.parallelize(values)
print(rdd2.collect())
````

````python
sc.stop()
````

````python
conf = SparkConf().setAppName("SparkRDDs").setMaster("local")
sc = SparkContext(conf=conf)
print("Hello, Spark!")
````

````python
values = [1, 2, 3, 4, 5]
rdd = sc.parallelize(values) # parallelize take collections like list etc.

# printing the RDD values
print(rdd.take(5))
````

````python
rdd = sc.parallelize([2,3,4])
sorted(rdd.flatMap(lambda x: range(1,x)).collect())
````
