This code distributes 4 elements across 3 partitions and computes the sum of each partition locally. The result is a list of partition sums returned to the driver.   
     
How elements are distributed in this case:

Partition 0: [1] → sum = 1.     
Partition 1: [2] → sum = 2.     
Partition 2: [3, 4] → sum = 7.

````python
x = sc.parallelize(range(0,3)).keyBy(lambda x: x*x)
y = sc.parallelize(zip(range(0,5),range(0,5)))
[(x,list(map(list,y))) for x,y in sorted(x.cogroup(y).collect())]
````

This code performs a cogroup operation between two RDDs (Resilient Distributed Datasets), grouping their values by key and collecting the results as a sorted list.     
      
Line 1: x = sc.parallelize(range(0,3)).keyBy(lambda x: x*x).   
    
Creates an RDD with values 0, 1, 2 using sc.parallelize().   
Transforms each element into a key-value pair using keyBy() where the key is the square of the value    
Result: [(0, 0), (1, 1), (2, 2)] — (key, value) pairs where key = value²
     
Line 2: y = sc.parallelize(zip(range(0,5),range(0,5)))     
     
Creates tuples (0,0), (1,1), (2,2), (3,3), (4,4) using zip(range(0,5), range(0,5))    
Parallelizes these tuples as an RDD     
Result: [(0, 0), (1, 1), (2, 2), (3, 3), (4, 4)] — (key, value) pairs    
        
Line 3: x.cogroup(y).collect()    
      
cogroup() groups values from both RDDs by their keys     
For each key, it produces an iterable of values from x and an iterable from y   
sorted() orders the results by key     
list(map(list, ...)) converts the iterables to lists.   
    
Expected Output.   
[(0, [[0], [0]]),    
 (1, [[1], [1]]),     
 (2, [[2], [2]]),      
 (3, [[], [3]]),     
 (4, [[], [4]])].   
     
This shows that keys 0, 1, 2 have values in both RDDs, while keys 3 and 4 only have values in RDD y.

````python
rdd = sc.textFile("Spark.txt")
nonempty_lines = rdd.filter(lambda x: len(x)>0)
words = nonempty_lines.flatMap(lambda x: x.split(' '))
wordcount = words.map(lambda x: (x,1)).reduceByKey(lambda x,y:x+y).map(lambda x:(x[1],x[0])).sortByKey(False)

for word in wordcount.collect():
	print(word)
````
