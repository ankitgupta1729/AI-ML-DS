```python
import time
```

```python
def brew_tea():
    print("Brewing Tea...")
    time.sleep(1)
    print("Tea is ready!")
brew_tea() # function is created for this purpose 
# and it will take approx 1 sec
```

```python
def brew_tea():
    start_time = time.time()
    print("Brewing Tea...")
    time.sleep(1)
    print("Tea is ready!")
    end_time = time.time()
    print(f"Task time: {end_time-start_time} seconds")
brew_tea() # function with additional feature to measure 
# exact time
```
